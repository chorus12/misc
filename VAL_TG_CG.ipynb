## Calculate Inertia for K-clusters  
below is a `python` code to generate sql with a query that calculates inertia for cluster distribution

```python
TABLE_NAME = "demo_db.ClusterInput" #  scaled table with data
CLUSTER_TABLE = "demo_db.ClusterScore" # all data scored to each cluster
CLUSTER_MODEL = "demo_db.ClusterModel" # cluster model metadata table
KEY = "cust_id" # key field in a table
CLUSTER_ID = "clusterid"

# fields description from DDL
s = '''
      tot_income DECIMAL(15,1),
      tot_age INTEGER,

      q2_trans_cnt INTEGER,
      q3_trans_cnt INTEGER,
      q4_trans_cnt INTEGER
'''

# extract all fields from fields definition above
fields = [f for i,f in enumerate(s.split()) if i%2 ==0]

sql_string = '''
with inertia_items as (
select 
  b.clusterid, 
'''
for f in fields:
    sql_string += f"  (a.{f} - c.{f})**2 + \n"
last_plus_pos = sql_string.find(" +", -4)
```

````
with inertia_items as (
select 
  b.clusterid, 
  (a.tot_income - c.tot_income)**2 + 
  (a.tot_age - c.tot_age)**2 + 
  (a.tot_cust_years - c.tot_cust_years)**2 + 
  (a.tot_children - c.tot_children)**2 + 
  (a.single_ind - c.single_ind)**2 + 
  (a.female_ind - c.female_ind)**2 + 
  (a.married_ind - c.married_ind)**2 + 
  (a.separated_ind - c.separated_ind)**2 + 
  (a.ca_resident_ind - c.ca_resident_ind)**2 + 
  (a.ny_resident_ind - c.ny_resident_ind)**2 + 
  (a.tx_resident_ind - c.tx_resident_ind)**2 + 
  (a.il_resident_ind - c.il_resident_ind)**2 + 
  (a.az_resident_ind - c.az_resident_ind)**2 + 
  (a.oh_resident_ind - c.oh_resident_ind)**2 + 
  (a.ck_acct_ind - c.ck_acct_ind)**2 + 
  (a.sv_acct_ind - c.sv_acct_ind)**2 + 
  (a.cc_acct_ind - c.cc_acct_ind)**2 + 
  (a.ck_avg_bal - c.ck_avg_bal)**2 + 
  (a.sv_avg_bal - c.sv_avg_bal)**2 + 
  (a.cc_avg_bal - c.cc_avg_bal)**2 + 
  (a.ck_avg_tran_amt - c.ck_avg_tran_amt)**2 + 
  (a.sv_avg_tran_amt - c.sv_avg_tran_amt)**2 + 
  (a.cc_avg_tran_amt - c.cc_avg_tran_amt)**2 + 
  (a.q1_trans_cnt - c.q1_trans_cnt)**2 + 
  (a.q2_trans_cnt - c.q2_trans_cnt)**2 + 
  (a.q3_trans_cnt - c.q3_trans_cnt)**2 + 
  (a.q4_trans_cnt - c.q4_trans_cnt)**2 as inertia
from 
  demo_db.ClusterInput a join 
  demo_db.ClusterScore b 
  on a.cust_id = b.cust_id join 
  demo_db.ClusterModel c 
  on c.clusterid = b.clusterid)
select CAST(sum(inertia) as int) inertia from inertia_items
````

````
REPLACE PROCEDURE demo_db.cluster_data(
    -- Procedure runs clustering over the data and calculates inertia for the clusterization
    OUT inertia BIGINT, -- the vale of inertia
    IN CLUSTER_NUM INTEGER,  -- number of clusters
    IN IN_DB varchar(100), -- database of incoming table
    IN MODEL_DB varchar(100), -- database of table with cluster model
    IN IN_TABLE_NAME varchar(100), -- input table with data to cluster
    IN COLS varchar(1000), -- columns of table to treat as dimentions
    IN MODEL_TABLE_NAME varchar(1000), -- table name of cluster model to store
    IN OPERATOR_DB varchar(100), -- database where VAL operator resides
    IN ITERATIONS INTEGER,  -- number of iterations in clustering 
    -- end of model parameters
    IN OUT_DB varchar(100), -- output database with clustered points
    IN OUT_TABLE_NAME varchar(100), -- output table name with clustered points
    IN KEY_COL_NAME varchar(100) -- key column that identifies objects - that will be in the output table
)
BEGIN
DECLARE cluster_param varchar(1000); -- param for clustering model
DECLARE run_cluster_param varchar(1000); --param for actual clustering
DECLARE i BIGINT;

--     'database=demo_db;
--     tablename=ClusterInput;
--     columns=tot_income,tot_age,tot_cust_years,
--     tot_children,single_ind,female_ind,married_ind,
--     separated_ind,ca_resident_ind,ny_resident_ind,
--     tx_resident_ind,il_resident_ind,az_resident_ind,
--     oh_resident_ind,ck_acct_ind,sv_acct_ind,cc_acct_ind,
--     ck_avg_bal,sv_avg_bal,cc_avg_bal,ck_avg_tran_amt,
--     sv_avg_tran_amt,cc_avg_tran_amt,
--     q1_trans_cnt,q2_trans_cnt,q3_trans_cnt,q4_trans_cnt;
--     outputdatabase=demo_db;
--     outputtablename=ClusterModel;
--     operatordatabase=TRNG_XSP;
--     kvalue=3;
--     iterations=10;
--     threshold=0.1;'

SET cluster_param='database='||IN_DB||';tablename='||IN_TABLE_NAME||';columns='||COLS;
SET cluster_param=cluster_param||';outputdatabase='||MODEL_DB||';outputtablename='||MODEL_TABLE_NAME||';operatordatabase='||OPERATOR_DB;
SET cluster_param=cluster_param||';kvalue='||CAST(CLUSTER_NUM as VARCHAR(3))||';iterations='||CAST(ITERATIONS as VARCHAR(3))||';threshold=0.001;';

-- create a cluster model
call TRNG_XSP.td_analyze(
    'Kmeans', cluster_param);
    
-- call TRNG_XSP.td_analyze(
--     'KmeansScore',
--     'database=demo_db;
--     tablename=ClusterInput;
--     outputdatabase=demo_db;
--     outputtablename=ClusterScore;
--     index=cust_id;
--     modeldatabase=demo_db;
--     modeltablename=ClusterModel;
--     operatordatabase=TRNG_XSP;');

-- run actual clustering 
SET run_cluster_param = 'database='||IN_DB||';tablename='||IN_TABLE_NAME||';outputdatabase='||OUT_DB;
SET run_cluster_param = run_cluster_param||';outputtablename='||OUT_TABLE_NAME||';index='||KEY_COL_NAME;
SET run_cluster_param = run_cluster_param||';modeldatabase='||MODEL_DB||';modeltablename='||MODEL_TABLE_NAME||';operatordatabase='||OPERATOR_DB||';';

call TRNG_XSP.td_analyze(
     'KmeansScore', run_cluster_param);
     
-- calculate inertia value
select CAST(sum(inertia) as int) into i from  (
select 
  b.clusterid, 
  (a.tot_income - c.tot_income)**2 + 
  (a.tot_age - c.tot_age)**2 + 
  (a.tot_cust_years - c.tot_cust_years)**2 + 
  (a.tot_children - c.tot_children)**2 + 
  (a.single_ind - c.single_ind)**2 + 
  (a.female_ind - c.female_ind)**2 + 
  (a.married_ind - c.married_ind)**2 + 
  (a.separated_ind - c.separated_ind)**2 + 
  (a.ca_resident_ind - c.ca_resident_ind)**2 + 
  (a.ny_resident_ind - c.ny_resident_ind)**2 + 
  (a.tx_resident_ind - c.tx_resident_ind)**2 + 
  (a.il_resident_ind - c.il_resident_ind)**2 + 
  (a.az_resident_ind - c.az_resident_ind)**2 + 
  (a.oh_resident_ind - c.oh_resident_ind)**2 + 
  (a.ck_acct_ind - c.ck_acct_ind)**2 + 
  (a.sv_acct_ind - c.sv_acct_ind)**2 + 
  (a.cc_acct_ind - c.cc_acct_ind)**2 + 
  (a.ck_avg_bal - c.ck_avg_bal)**2 + 
  (a.sv_avg_bal - c.sv_avg_bal)**2 + 
  (a.cc_avg_bal - c.cc_avg_bal)**2 + 
  (a.ck_avg_tran_amt - c.ck_avg_tran_amt)**2 + 
  (a.sv_avg_tran_amt - c.sv_avg_tran_amt)**2 + 
  (a.cc_avg_tran_amt - c.cc_avg_tran_amt)**2 + 
  (a.q1_trans_cnt - c.q1_trans_cnt)**2 + 
  (a.q2_trans_cnt - c.q2_trans_cnt)**2 + 
  (a.q3_trans_cnt - c.q3_trans_cnt)**2 + 
  (a.q4_trans_cnt - c.q4_trans_cnt)**2 as inertia
from 
  demo_db.ClusterInput a join 
  demo_db.ClusterScore b 
  on a.cust_id = b.cust_id join 
  demo_db.ClusterModel c 
  on c.clusterid = b.clusterid) ttt;

SET inertia = i;

END;
````

````
REPLACE PROCEDURE demo_db.cluster_data_multiple(
    --///////////////////////////////////////////////////////////////////////////////////////////////////////
    -- Procedure calculates inertia for clustering data from strat_clusters to max_clusters
    -- results are stored in volatile table cluster_quality (cluster_num, inertia)
    -- Next step this table is used to define optimal clusters number
    --///////////////////////////////////////////////////////////////////////////////////////////////////////
    
    IN start_clusters Integer,
    IN max_clusters INTEGER,
    -- Procedure runs clustering over the data and calculates inertia for the clusterization
    IN IN_DB varchar(100), -- database of incoming table
    IN MODEL_DB varchar(100), -- database of table with cluster model
    IN IN_TABLE_NAME varchar(100), -- input table with data to cluster
    IN COLS varchar(1000), -- columns of table to treat as dimentions
    IN MODEL_TABLE_NAME varchar(1000), -- table name of cluster model to store
    IN OPERATOR_DB varchar(100), -- database where VAL operator resides
    IN ITERATIONS INTEGER,  -- number of iterations in clustering 
    -- end of model parameters
    IN OUT_DB varchar(100), -- output database with clustered points
    IN OUT_TABLE_NAME varchar(100), -- output table name with clustered points
    IN KEY_COL_NAME varchar(100) -- key column that identifies objects - that will be in the output table
)
SQL Security invoker
BEGIN
  DECLARE k INT DEFAULT 2; -- current number of clusters
  DECLARE current_inertia BIGINT; -- inertia for current K number of clusters

  CALL demo_db.drop_my_table('cluster_quality'); -- drop table if it exists before creating
  
  -- table to store inertia for all clusters
  CREATE MULTISET VOLATILE TABLE cluster_quality (cluster_num SMALLINT, inertia BIGINT) 
  PRIMARY INDEX ( cluster_num ) ON COMMIT PRESERVE ROWS;
  
  SET k = start_clusters;
  -- main loop to create cluster models and calculate inertia
  WHILE k < max_clusters+1
  DO
    call demo_db.cluster_data(current_inertia, k, IN_DB, MODEL_DB, IN_TABLE_NAME, 
        COLS, MODEL_TABLE_NAME, OPERATOR_DB, ITERATIONS, OUT_DB, OUT_TABLE_NAME, KEY_COL_NAME);
    insert into cluster_quality (cluster_num, inertia) values (k, current_inertia);
    SET k = k+1;
  END WHILE;

END;
````

### This is a thrilling but yet quite simple query that tries to implement an elbow method for definition of number of clusters  
Main idea is that rate of decrease for inertia decreases to some point and then becomes stable.  
This you can see on smoothed inertia diff (see picture **Normalized smoothed inertia diff** below)  

````
replace view demo_db.v_elbow as 
-- single inertia diff
with inertia_filtered as (
select 
  cluster_num, 
  inertia, 
  inertia_diff
from (
select cluster_num, inertia, -mdiff(inertia, 1, cluster_num) as inertia_diff  
from cluster_quality 
qualify coalesce(inertia_diff, 1) > 0) a ), 

-- remove points with increasing inertia
inertia_filtered_diff as (
select  
  cluster_num, 
  inertia, 
  -mdiff(inertia, 1, cluster_num) as inertia_diff
from inertia_filtered), 

-- min-max scale inertia diff
inertia_normalized as (
select 
  cluster_num, 
  cast((inertia_diff - min_in_diff)/(max_in_diff - min_in_diff) as decimal(5,3)) as normed_diff,
  inertia, 
  inertia_diff, 
  cast(min(inertia_diff) over () as float) as min_in_diff,
  max(inertia_diff) over () as max_in_diff
from inertia_filtered_diff
      where inertia_diff is not null), 

-- smoothing inertia diff
smoothed_normed_diff as (
select 
  cluster_num, 
  normed_diff, 
  cast(avg(normed_diff) over (order by cluster_num rows between 3 PRECEDING and 0 preceding) as decimal(3,2)) as smoothed_diff
from inertia_normalized), 

-- adding double diff for smoothed_diff      
smoothed_normed_double_diff as (
select 
      cluster_num, normed_diff, smoothed_diff, -mdiff(smoothed_diff, 1, cluster_num) as double_diff
from smoothed_normed_diff),
      
-- adding smoothing for double_diff
double_smoothed_double_diff as (
select 
      cluster_num, normed_diff, double_diff, smoothed_diff,
      cast(avg(double_diff) over (order by cluster_num rows between 3 PRECEDING and 0 preceding) as decimal(5,2)) as smoothed_double_diff
from smoothed_normed_double_diff),

-- and a final part adding a stop_flag - this is a minimum of smoothed_double_diff from current point and two points ahead
-- if theis value is lower then a thereshold (a value close to zero) - it is an elbow point
elbow_view as (
select
  cluster_num, 
  normed_diff, 
  smoothed_diff, 
  double_diff,
  smoothed_double_diff,
  min(smoothed_double_diff) over (order by cluster_num rows between 0 preceding and 2 following) as stop_flag
from double_smoothed_double_diff)

select * from elbow_view;
````

## Create a procedure for TG/CG split  


````
REPLACE PROCEDURE demo_db.tgcg_split(
    -- Procedure runs actual TG CG split
    IN cg_ratio decimal(3,2), -- fraction of data that goes into Control Group
    IN db_name varchar(100), -- database to hold ClusterScore table (not used at the moment!!!)
    IN cluster_score_table_name varchar(100), -- ClusterScore table name (not used at the moment!!!)
    -- need to make dynamic cusror - if needed
    OUT test varchar(1000)
) 
BEGIN
  DECLARE sql_text varchar(1000);
    
  -- drop table for TG-CG split if it exists
  call demo_db.drop_my_table('demo_db.tg_cg_split');
  -- create table to hold TG-CG split
  CREATE TABLE demo_db.tg_cg_split (
      gr varchar(2),
      clusterid INTEGER,
      cust_id INTEGER) 
  primary index (cust_id);
  
  
  -- main loop over the clusters
  FOR clusters AS cur1 CURSOR FOR 
    SELECT  distinct clusterid 
    FROM  demo_db.clustermodel 
    where clusterid > 0 order by 1
  DO
    --the below code does not work in TSQL - so we make it dynamic
    --insert into demo_db.tg_cg_split select 'CG', clusterid, cust_id from demo_db.ClusterScore where clusterid = 1 sample RANDOMIZED ALLOCATION 0.05
    -- we use smaple clause to select cg_ratio of population as Control Group for the cluster 
    set sql_text = 'insert into demo_db.tg_cg_split 
    select ''CG'', clusterid, cust_id  from demo_db.ClusterScore 
    where clusterid = '||CAST(clusters.clusterid as char(3))||' sample RANDOMIZED ALLOCATION ' || CAST (cg_ratio as char(5))||';'; 
    CALL DBC.SysExecSQL(sql_text);
        
    -- remaining rows belong to TG and we insert them appropriately
    insert into demo_db.tg_cg_split 
    select 'TG', clusterid, cust_id from demo_db.ClusterScore 
    where 
      clusterid = clusters.clusterid and 
      cust_id not in (select cust_id from demo_db.tg_cg_split);  
  END FOR;
  set test = sql_text;
  
END;
````

````
-- display the stats
with tg_cg_stat as (
select gr, 
  clusterid, 
  count(distinct cust_id) cnt,
  max(cnt) over (partition by clusterid) as tg_cnt,
  min(cnt) over (partition by clusterid) as cg_cnt,
  cast(cast(cg_cnt as float)/(cg_cnt+tg_cnt)  as decimal(3,2)) as cg_ratio
from demo_db.tg_cg_split group by gr, clusterid)
select clusterid, tg_cnt, cg_cnt, cg_ratio from tg_cg_stat where gr = 'CG' order by clusterid;
````
